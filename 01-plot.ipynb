- learn how to use SigPy plotting functions.

https://sigpy.readthedocs.io/en/latest/plot.html

```python
%matplotlib notebook
import numpy as np
import sigpy as sp
import sigpy.plot as pl
```

# Image Plots

```python
x = sp.shepp_logan([64, 64, 64])
pl.ImagePlot(x)
```

# Line Plots

```python
pl.LinePlot(x)
```

# Scatter Plots

```python
N = 100
values = np.arange(N)

x = np.arange(N)
y = np.arange(N)
coord = np.stack([x, y], axis=1)

pl.ScatterPlot(coord, x)
```
